```python
#Import dependencies

import pandas as pd
import numpy as ny
```

```python
#Import json file, print first five

json_path = "purchase_data.json"
Game_Data_df = pd.read_json(json_path)

Game_Data_df.head()

```

```python
#Total number of players output

total_players = Game_Data_df["SN"].nunique() 
players = total_players

print('Total Players', players)

```

```python
#Purchasing Analysis

#Number of Unique Items 

Unique_Items_df = Game_Data_df['Item Name'].nunique(dropna=True)
print('Number of Unique Items', Unique_Items_df)


#Average Purchase Price 
average = Game_Data_df["Price"].mean()
Average = round(average,2)
print('Average', Average)

#Total Number of Purchases
Total_Purchases = Game_Data_df['Item ID'].count()
print('Total Purchases', Total_Purchases)  

#Total Revenue
Total_Revenue = Game_Data_df['Price'].sum()
print('Total Revenue','$',Total_Revenue)
```

```python
#Gender Demographics

#Number of Male Players
Males = Game_Data_df.loc[Game_Data_df.Gender=='Male', 'SN']
Number_Males = Males.nunique()
print("Total Male Players", Number_Males)


#Number of Female Players
Females = Game_Data_df.loc[Game_Data_df.Gender=='Female', 'SN']
Number_Females = Females.nunique()
print("Total Female Players", Number_Females)  

#Number Non-Disclosed
Other = Game_Data_df.loc[Game_Data_df.Gender=='Other / Non-Disclosed', 'SN']
Number_Others = Other.nunique()   
print("Total Non-Disclosed", Number_Others)
```

```python
#Purchasing Analysis, Gener

# The below each broken by gender, Purchase Count, Average Purchase Price, Total Purchase Value, Normalized Totals

F = Game_Data_df.loc[Game_Data_df.Gender=='Female', 'Price']

F_P = F.count()
F_M = F.mean()
F_m = round(F_M,2)
F_S = F.sum()
F_s = round(F_S,2)
F_N_T = F_S/F_P  
F_N_t = round(F_N_T,2)

#normalized_total = gender_purchase_total / gender_demographics["Total Count"]

U = Game_Data_df.loc[Game_Data_df.Gender=='Other / Non-Disclosed', 'Price']
U_P = U.count()
U_M = U.mean()
U_m = round(U_M,2)
U_S = U.sum()
U_s = round(U_S,2)
U_N_T = U_S/U_P  
U_N_t = round(U_N_T,2)

M = Game_Data_df.loc[Game_Data_df.Gender=='Male', 'Price']

M_P = M.count()
M_M = M.mean()
M_m = round(M_M,2)
M_S = M.sum()
M_s = round(M_S,2)
M_N_T = M_S/M_P  
M_N_t = round(M_N_T,2)

print('Female:  Purchase Count',F_P, 'Average Purchase Price','$',F_m,'Total Purchase Value', '$',F_s, 'Normalized Value', '$', F_N_t)
print('Undisclosed:  Purchase Count',U_P, 'Average Purchase Price','$',U_m,'Total Purchase Value', '$',U_s, 'Normalized Value', '$', U_N_t)
print('Male: Purchase Count',M_P, 'Average Purchase Price','$',M_m,'Total Purchase Value', '$',M_s, 'Normalized Value', '$', M_N_t)


```

```python
#Age Demographics

bins = [10, 15, 20, 25, 30, 35, 40]

Game_Data_df["Age"] = pd.cut(Game_Data_df["Age"], bins)

Game = Game_Data_df.groupby('Age')

Count = (Game['Price'].count())
Mean = (Game['Price'].mean())
mean = round(Mean,2)
Sum = (Game['Price'].sum())


Age_Dataframe = pd.DataFrame({'Count': Count, 'Average': mean, "Sum": Sum})

print(Age_Dataframe)
```

```python
#Identify the the top 5 spenders in the game by total purchase value, then list (in a table):
#SN, Purchase Count, Average Purchase Price, Total Purchase Value
#ted_group = ted_df.groupby("View Group")


User_Total = Game_Data_df.groupby(["SN"]).sum()["Price"].rename("Total Purchase Value")

User_Average = Game_Data_df.groupby(["SN"]).mean()['Price'].rename('Average_Purchase_value')

User_Purchase = Game_Data_df.groupby(["SN"]).count()['Price'].rename('Count_Items')

df = pd.DataFrame({'Total Purchase' : User_Total, 'User Averavge': User_Average, 'User Purchase': User_Purchase})

Last_df = df.sort_values('Total Purchase', ascending=False)

Last_df.head()



```

```python

#Identify the 5 most popular items by purchase count, then list (in a table):
#Item ID, Item Name, Purchase Count, Item Price, Total Purchase Value
#Purchase Count

Popular_Item = Game_Data_df.groupby(['Item ID']).count()['Price'].rename('Most_Popular')
Purchase_Price = Game_Data_df.groupby(['Item ID']).sum()['Price'].rename('Total_Purchase_value')
df = pd.DataFrame({'Total Purchase' : Purchase_Price, 'User Purchase': Popular_Item,})
Last_DF = df.sort_values('User Purchase', ascending=False)
Last_DF.head()
```

```python
#Identify the 5 most profitable items by total purchase value, then list (in a table):
#Item ID, Item Name, Purchase Count, Item Price, Total Purchase Value

Popular_Item = Game_Data_df.groupby(['Item ID']).count()['Price'].rename('Most_Popular')
Purchase_Price = Game_Data_df.groupby(['Item ID']).sum()['Price'].rename('Total_Purchase_value')
df = pd.DataFrame({'Total Purchase' : Purchase_Price, 'User Purchase': Popular_Item})
Last_DF = df.sort_values('User Purchase', ascending=False)
Last_DF.head()
```
